```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
deps = pd.read_excel('C:\\Users\\C23Daniel.Joseph\\Downloads\\2022_BAH_Rates.xls', header = 0,sheet_name = 'With')
noDeps = pd.read_excel('C:\\Users\\C23Daniel.Joseph\\Downloads\\2022_BAH_Rates.xls', header = 0, index_col = 0,sheet_name = 'Without')
difference = pd.read_excel('C:\\Users\\C23Daniel.Joseph\\Downloads\\2022_BAH_Rates.xls', header = 0, index_col = 0,sheet_name = 'Difference')
```

```python
depsMelted = pd.melt(deps, id_vars=['MHA','MHA_NAME'])
depsMelted.MHA_NAME = depsMelted.MHA_NAME.str.strip()
```

```python
depsMelted
```

```python
fig1 = px.line(depsMelted[depsMelted.variable.str.startswith('E')],
               x = depsMelted[depsMelted.variable.str.startswith('E')].variable,
               y = depsMelted[depsMelted.variable.str.startswith('E')].value,
               color ='MHA_NAME',
               labels = {
                   'x': 'Rank (NATO Scale)',
                   'y': 'BAH ($)',
                   'MHA_NAME':'Location'
               },
               title="Enlisted BAH Pay w/ Dependents<br><sup>Data from Wherever Nico Got It </sup>")


fig1.show()
```

```python
noE = depsMelted[~depsMelted.variable.str.contains('E')]
onlyO = noE[noE.variable.str.startswith('')]

fig2 = px.line(onlyO,
               x = onlyO.variable,
               y = onlyO.value,
               color ='MHA_NAME',
               labels = {
                   'x': 'Rank (NATO Scale)',
                   'y': 'BAH ($)',
                   'MHA_NAME':'Location'
               },
               markers = True,
               title="Officer BAH Pay w/ Dependents<br><sup>Data from Wherever Nico Got It </sup>")



fig2.show()
```

```python
depsRatio = deps
depsRatio.drop(['E08', 'E09', 'Unnamed: 11', 'W01', 'W02', 'W03', 'W04', 'W05', 'Unnamed: 17', 'O01E', 'O02E', 'O03E', 'Unnamed: 21'], axis = 1, inplace = True)
depsRatio['O1:E1'] = depsRatio.O01/depsRatio.E01
depsRatio['O2:E2'] = depsRatio.O02/depsRatio.E02
depsRatio['O3:E3'] = depsRatio.O03/depsRatio.E03
depsRatio['O4:E4'] = depsRatio.O04/depsRatio.E04
depsRatio['O5:E5'] = depsRatio.O05/depsRatio.E05
depsRatio['O6:E6'] = depsRatio.O06/depsRatio.E06
depsRatio['O7:E7'] = depsRatio.O07/depsRatio.E07

depsRatio.drop(['E01', 'E02', 'E03', 'E04', 'E05', 'E06', 'E07',
       'O01', 'O02', 'O03', 'O04', 'O05', 'O06', 'O07'], axis = 1, inplace = True)
```

```python
depsRatioMelted = pd.melt(depsRatio, id_vars=['MHA','MHA_NAME'])
depsRatio
```

```python
fig3 = px.line(depsRatioMelted,
               x = depsRatioMelted.variable,
               y = depsRatioMelted.value,
               color ='MHA_NAME',
               labels = {
                   'variable': 'Rank',
                   'value': 'Ratio',
                   'MHA_NAME':'Location'
               },
               #markers = True,
               title="Ratio of Officer BAH to Enlisted BAH (w/ Dependents)<br><sup>Data from Wherever Nico Got It </sup>")



fig3.show()
```
